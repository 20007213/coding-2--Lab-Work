### Load and Crop

```python
import matplotlib.pyplot as plt
```

```python
image = plt.imread('01.png')
plt.imshow(image)
print(image.shape)
```

```python
h = image.shape[0]
w = image.shape[1]
#image = image.crop(((w-h)/2),0,(w-h)/2+h,h)[:,:,0]
image_cropped = image[:,int((w-h)/2):int((w-h)/2+h),0]
print(image_cropped.shape)
plt.figure(figsize=(9,9))
plt.axis('off')
plt.imshow(image_cropped,cmap='gray')
```

```python
import numpy as np
```

```python
area = np.ma.masked_where(image_cropped < 0.5, image_cropped)
```

```python
area.mask
```

```python
plt.imshow(area.mask, cmap='gray')
```

```python
np.random.seed(19680801)

N = 7000
x = np.random.rand(N)*1158
y = np.random.rand(N)*1158

c = np.sqrt(area)

plt.figure(figsize=(9,9))
for i in range(N):
    if area.mask[int(x[i]),int(y[i])]:
        plt.scatter(y[i],x[i],  marker='o').axes.invert_yaxis()
plt.show()
```
